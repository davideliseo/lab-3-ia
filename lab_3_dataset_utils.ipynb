# Utilidades para la adaptación del dataset

```python
from google.colab import drive
import os

# Monta la unidad de Google Drive
drive.mount("/content/drive", force_remount=True)
```

```python
# Ruta base de la aplicación.
path = "/content/drive/Shareddrives/IA/Lab 3"

# Ruta del dataset
dataset_path = os.path.join(path, "dataset")

# Ruta del dataset exportado
dataset_backup_path = os.path.join(path, "backup/dataset_08062023")

revised_dataset_path = os.path.join(path, "dataset_rev")
```

```python
# Clases del dataset
_, dirs, _ = next(os.walk(dataset_path))
class_names = dirs
```

```python
import pathlib

def unique_dataset_extensions():
  return set(
    "".join(path.suffixes).lower() 
    for path in pathlib.Path(revised_dataset_path).rglob("*.*")
  )

unique_dataset_extensions()
```

```python
from PIL import Image
import os

i = 1

def image_to_jpeg(source_filepath, target_filepath):
  try:
    # Crear directorios recursivamente si no existen
    os.makedirs(os.path.dirname(target_filepath), exist_ok=True)

    # Abre la imagen, la convierte en una imagen de 3 canales y la guarda como JPEG
    image = Image.open(source_filepath).convert("RGB")
    image.save(target_filepath, "JPEG")

    print(f"Imagen guardada: {target_filepath}")
  except Exception as e:
    print(f"Error al convertir el archivo {source_filepath}: {e}")


def images_to_jpeg(source_directory, target_directory, class_name):
    global i

    # Recorrer cada archivo en el directorio de origen
    for source_filename in os.listdir(source_directory):
      source_filepath = os.path.join(source_directory, source_filename)

      if os.path.isfile(source_filepath):
        target_filename, _ = os.path.splitext(os.path.join(target_directory, f"{class_name}_{i}"))
        target_filepath = f"{target_filename}.jpeg"
        image_to_jpeg(source_filepath, target_filepath)
        i += 1


def dataset_to_jpeg(class_names=class_names):
  global i
  for class_name in class_names:
    i = 1
    images_to_jpeg(source_directory=os.path.join(revised_dataset_path, class_name), 
                   target_directory=os.path.join(dataset_path, class_name),
                   class_name=class_name)
```

```python
dataset_to_jpeg(["paint_brush", "shovel"])
```
